
Problem 5

```python
#Problem 5
min_age = table["age"].min()
max_age = table["age"].max()

min_income = table["income"].min()
max_income = table["income"].max()

min_children = table["children"].min()
max_children = table["children"].max()

print(min_age, max_age, min_income, max_income, min_children, max_children)
```

```python
norm_income = (table["income"] - min_income) / (max_income-min_income)
```

```python
norm_age = (table["age"] - min_age) / (max_age-min_age)
```

```python
norm_children = (table["children"] - min_children) / (max_children-min_children)
```

```python
table["income"] = norm_income
table["age"] = norm_age
table["children"] = norm_children
```
